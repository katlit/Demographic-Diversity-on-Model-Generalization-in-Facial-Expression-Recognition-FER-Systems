```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
```

```python
annot1_data = pd.read_csv('annotations_1.csv')
annot2_data = pd.read_csv('annotations_2.csv')
annot3_data = pd.read_csv('annotations_3.csv')
#selection_data = pd.read_csv('selection.csv')
```

```python
# Merge all annotation datasets with the updated selection data
merged_data = annot1_data.merge(annot2_data, on='Image', how='outer', suffixes=('_A1', '_A2'))
merged_data = merged_data.merge(annot3_data, on='Image', how='outer', suffixes=(None, '_A3'))

# Fill missing columns with default values
#columns_to_fill = [col for col in merged_data.columns if '_A1' in col or '_A2' in col or '_A3']
#merged_data[columns_to_fill] = merged_data[columns_to_fill].fillna("NaN")  # Replace with "NaN" or other default value

# Drop duplicate "emotions" columns
merged_data = merged_data.drop(columns=['Emotion_A1','Annotation_by_A1', 
                                        'Emotion_A2','Annotation_by_A2', 
                                        'Annotation_by'])

# Rename specific columns with suffixes
merged_data = merged_data.rename(columns={
    'Age': 'Age_A3',
    'Gender': 'Gender_A3',
    'Ethnicity': 'Ethnicity_A3',
    'Emotion': 'FolderNames'
})


# Display the merged data
merged_data.head()
```

```python
merged_data.columns
```

```python
# Rearrange columns in the desired order
desired_column_order = [
    'Image', 'FolderNames', 
    'Age_A1', 'Gender_A1', 'Ethnicity_A1',
    'Age_A2', 'Gender_A2', 'Ethnicity_A2',
    'Age_A3', 'Gender_A3', 'Ethnicity_A3',
]


# Ensure all columns in the desired order are present in the merged data
# Fill missing columns to avoid KeyError if any column is not present in the merged data
for col in desired_column_order:
    if col not in merged_data.columns:
        merged_data[col] = "NaN"  # Fill with default value if missing

# Reorder the columns
merged_data = merged_data[desired_column_order]

# Save the reordered data as a CSV file
merged_data.to_csv('annotationsManual_ALL.csv', index=False)

# Display the reordered data
merged_data
```
